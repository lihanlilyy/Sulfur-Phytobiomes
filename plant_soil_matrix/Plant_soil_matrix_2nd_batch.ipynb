```python
import csv
import textwrap
import os
import collections
import pandas as pd
pd.options.mode.chained_assignment = None
```

```python
folder_to_view = "fullset"
df = pd.read_csv('/Users/hanli/Desktop/FYP/hmmscan/subset/Score.csv')
print(df)
```

```python
for i, file in enumerate(os.listdir(folder_to_view)):
    if file.endswith(".tsv"):
        print(file)
        df[file] = ""
        for row_num in range (0, 177):
            df.loc[row_num, file] = 0
        directory = os.path.dirname(os.path.abspath(file))
        print(file + " is done")
```

```python
df.to_csv("test.csv")
```

```python
for i, file in enumerate(os.listdir(folder_to_view)):
    if file.endswith(".tsv"):
        print(file)
        directory = os.path.dirname(os.path.abspath(file))
        with open(directory+"/fullset/"+file, 'rt') as f:
            for row in csv.reader(f, delimiter='\t'):
                row = row[0].split(" ")
                new_row = [x for x in row if x != '']
                fullname = str(new_row[0])
                firstLetter = new_row[0][0]
                if (firstLetter != "#") & (fullname in df['Gene'].values):
                    standard_score = df.loc[df['Gene'] == str(fullname), 'Score'].iloc[0]
                    if float(new_row[5]) >= float(standard_score) : 
                        df.loc[df['Gene'] == fullname, file] += 1
        print(file + " is done")
```
